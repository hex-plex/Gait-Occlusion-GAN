```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import keras
from scipy.stats import norm
from keras import backend as K
```

```python
(x_tr, y_tr) , (x_te, y_te) = keras.datasets.mnist.load_data()
x_tr, x_te = x_tr.astype('float32')/255., x_te.astype('float32')/255.
x_tr, x_te = x_tr.reshape(x_tr.shape[0], -1), x_te.reshape(x_te.shape[0], -1)
print(x_tr.shape, x_te.shape)

y_tr, y_te = keras.utils.to_categorical(y_tr), keras.utils.to_categorical(y_te)
print(y_tr.shape, y_te.shape)

batch_size, n_epoch = 50, 50
n_hidden, z_dim = 512, 2
```

```python
x = keras.layers.Input(shape=(x_tr.shape[1:]))
condition = keras.layers.Input(shape=(y_tr.shape[1],))

inputs = keras.layers.concatenate([x, condition])
x_encoded = keras.layers.Dense(n_hidden, activation='relu')(inputs)
x_encoded = keras.layers.Dense(n_hidden//2, activation='relu')(x_encoded)
mu = keras.layers.Dense(z_dim, activation='linear')(x_encoded)
log_var = keras.layers.Dense(z_dim, activation='linear')(x_encoded)
```

```python
def sampling(args):
    mu, log_var = args
    eps = K.random_normal(shape=(batch_size, z_dim), mean=0., stddev=1.0)
    return mu + K.exp(log_var/2.)*eps
z = keras.layers.Lambda(sampling, output_shape=(z_dim,))([mu,log_var])
z_cond = keras.layers.concatenate([z, condition])
```

```python
z_decoder1 = keras.layers.Dense(n_hidden//2, activation='relu')
z_decoder2 = keras.layers.Dense(n_hidden, activation='relu')
y_decoder = keras.layers.Dense(x_tr.shape[1], activation='sigmoid')

z_decoded = z_decoder1(z_cond)
z_decoded = z_decoder2(z_decoded)
y = y_decoder(z_decoded)
```

```python
reconstruction_loss = keras.losses.binary_crossentropy(x,y)*x_tr.shape[1]
kl_loss = 0.5*K.sum(K.square(mu) + K.exp(log_var) - log_var - 1, axis=-1)
cvae_loss = reconstruction_loss + kl_loss

cvae = keras.models.Model([x, condition], y)
cvae.add_loss(cvae_loss)
cvae.compile(optimizer='adam',loss=None)
cvae.summary()
```

```python
cvae.fit([x_tr, y_tr],
            shuffle=True,
            epochs=n_epoch,
            batch_size=batch_size,
            validation_data=([x_te, y_te], None),
        )
```

```python
encoder = keras.models.Model([x,condition],mu)
encoder.summary()
```

```python
te_latent = encoder.predict([x_te, y_te], batch_size=batch_size)
plt.figure(figsize=(6,6))
plt.scatter(te_latent[:,0],te_latent[:,1], c=np.argmax(y_te, axis=1))
plt.colorbar()
plt.show()
```

```python
decoder_input = keras.layers.Input(shape=(z_dim+y_tr.shape[1],))
_z_decoded = z_decoder1(decoder_input)
_z_decoded = z_decoder2(_z_decoded)
_y = y_decoder(_z_decoded)
generator = keras.models.Model(decoder_input, _y)
generator.summary()
```

```python
digit_size = 28
for cond_num in range(10):
    condition_num = keras.utils.to_categorical(cond_num, 10).reshape(1, -1)
    plt.figure(figsize=(20,2))
    
    for i in range(10):
        z_sample = np.array([[0.3*i, 0.3]])
        x_decoded = generator.predict(np.column_stack([z_sample, condition_num]))
        
        digit = x_decoded[0].reshape(digit_size, digit_size)
        
        plt.subplot(1, 10, i+1)
        plt.axis('off')
        plt.imshow(digit, cmap='Greys_r')
        
    plt.show()
```
